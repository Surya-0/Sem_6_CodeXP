## Line plots

### Opening stock value

```python
plt.plot(df['Open'])
plt.xlabel('Timeline')
plt.ylabel('Stock value')
plt.show()
```

### Closing stock value

```python
plt.plot(df['Close'])
plt.xlabel('Timeline')
plt.ylabel('Stock value')
plt.show()
```

### High stock value

```python
plt.plot(df['High'])
plt.xlabel('Timeline')
plt.ylabel('Stock value')
plt.show()
```

### Low stock value

```python
plt.plot(df['Low'])
plt.xlabel('Timeline')
plt.ylabel('Stock value')
plt.show()
```
