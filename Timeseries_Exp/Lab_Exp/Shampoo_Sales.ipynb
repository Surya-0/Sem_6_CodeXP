```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from  datetime import datetime
```

```python
df = pd.read_csv('shampoo.csv')
df
```

```python
def parser(x):
	return datetime.strptime('190'+x, '%Y-%m')
```

```python
series = pd.read_csv('shampoo.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)
```

```python
series
```

```python
series.plot()
plt.show()
```

```python
upsampled = series.resample('D')
```

```python
upsampled.mean()
```

```python
interpolate = upsampled.interpolate(method= 'linear')
```

```python
interpolate
```

```python
interpolate.plot()
plt.show()
```

```python
interpolate2 = upsampled.interpolate(method= 'spline',order = 2) # Quadratic interpolation
interpolate2
```

```python
interpolate2.plot()
plt.show()
```

```python
down_sample = series.resample('Q')
down_sample.mean()
```

```python
quarterly_mean_sales = down_sample.mean()
quarterly_mean_sales.plot()
plt.show()
```

```python
resample = series.resample('A')
quarterly_mean_sales = resample.sum()
print(quarterly_mean_sales.head())
quarterly_mean_sales.plot()
plt.show()
```
